```python
df = {
    'Employee_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Edward'],
    'Department': ['HR', 'IT', 'IT', 'Marketing', 'Sales'],
    'Age': [29, 34, 41, 28, 38],
    'Salary': [50000, 70000, 65000, 55000, 60000],
    'Years_of_Experience': [4, 8, 10, 3, 12],
    'Joining_Date': ['2020-03-15', '2017-07-19', '2013-06-01', '2021-02-10', '2010-11-25'],
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Male'],
    'Bonus': [5000, 7000, 6000, 4500, 5000],
    'Rating': [4.5, 4.0, 3.8, 4.7, 3.5]
}
df=pd.DataFrame(df)
df
```

```python
print(np.mean(df['Salary']))
print(np.sum(df['Bonus']))
print(df['Age'].min())
print(df['Rating'].max())
```

```python
df_mod=df.sort_values(by='Salary',ascending=False)
df_mod
```

```python
def categorize(rating):
    if rating >= 4.5:
        return "Excellent"
    elif rating >= 4.0:
        return "Good"
    else:
        return "Average"

df['Performance_Category'] = df['Rating'].apply(categorize)
df
```

```python
print(df.isnull().sum())
```

```python
df.rename(columns={'Employee_ID': 'ID'}, inplace=True)
df
```

```python
grouped_by_experience = df.groupby(df['Years_of_Experience'] > 5)
more_than_5_years = grouped_by_experience.get_group(True)
more_than_5_years
```

```python
grouped_by_dept = df.groupby(df['Department'] == 'IT')
is_IT = grouped_by_dept.get_group(True)
is_IT
```

```python
df['Tax'] = df['Salary'] * 0.1
df['Salary'] = df['Salary'] * 0.9
df
```

```python
df.to_csv("modified_employees.csv", index=False)
```
